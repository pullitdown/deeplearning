```python
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l

true_w=torch.tensor([2,-3.4,10])
true_b=4.2
features,labels=d2l.synthetic_data(true_w,true_b,1000)
```

```python
def load_array(data_arrays,batch_size,is_train=True):
    dataset=data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset,batch_size,shuffle=is_train)

batch_size=10
data_iter=load_array((features,labels),batch_size)
```

```python
next(iter(data_iter))
```

```python
from torch import nn
net=nn.Sequential(nn.Linear(3,1))#决定输入特征和输出特征的形状,都是n*1
```

```python
net[0].weight.data.normal_(0,0.01)#因为均值为0,方差为0.01,说明生成的权重几乎全为0
print(net[0].weight.data)
net[0].bias.data.fill_(0)
```

```python
#定义损失函数
loss=nn.MSELoss(reduction='mean')
trainer=torch.optim.SGD(net.parameters(),lr=0.03)#管理了net的所有参数和梯度
```

```python
num_epoch=5

for i in range(num_epoch):
    for x,y in data_iter:
        # print("1  ",net[0].weight.grad)
        trainer.zero_grad()#因为要在backward之前把上一轮的梯度给去掉
        # print("2  ",net[0].weight.grad)
        l=loss(net(x),y)#l是矩阵,如果loss改为nn.MSELoss(reduction="sum"),那么返回的是一个标量
        print(l)
        # print("3  ",net[0].weight.grad)        
        l.backward()#只有调用了backward(),优化器所管理的参数的梯度才会改变
        # print("4  ",net[0].weight.grad)
        trainer.step()#都是小批量随机梯度下降
        # print("5  ",net[0].weight.grad)
    l=loss(net(features),labels)
    print(f"epoch{i+1},loss{l:.5f}")
```
